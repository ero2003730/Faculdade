```python
import pandas as pd
import zipfile
import numpy as np
```

```python
zf = zipfile.ZipFile('/content/drive/MyDrive/Diabetes_Classificacao/diabetes-classification.zip')  #aloca o arquivo zip

data_train = pd.read_csv(zf.open('train.csv')) # abre o arquivo CSV 'train.csv' presente dentro do ZIP


data_test = pd.read_csv('/content/drive/MyDrive/Diabetes_Classificacao/new_test.csv')   # abre o arquivo CSV 'new_test.csv' que foi criado após adição da coluna 'diabetes_svm'
```

```python
import sklearn

data_train = data_train.drop('p_id', axis=1) # tirar a coluna ID, pois não importa

data_train = sklearn.utils.shuffle(data_train) # embaralhar as linhas, para assim pegar os 10%
```

```python
x_train1 = data_train.values[ : 61, :8] # pegando 0 - 10% as 8 primeiras colunas
y_train1 = data_train.values[ : 61, 8] # pegando 0 - 10% a última coluna
```

```python
x_test = data_test.values[ : , : 8] # pegar o teste x
y_test = data_test.values[ : , 8] # pegar o teste y
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import balanced_accuracy_score

clf = DecisionTreeClassifier(random_state=0, max_depth=2)
clf.fit(x_train1, y_train1)
predicao = clf.predict(x_test)
ConfusionMatrixDisplay.from_estimator(clf, x_test, y_test)

acu1 = balanced_accuracy_score(y_test, predicao)
print(acu1)
```
